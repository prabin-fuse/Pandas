## Indexing in panads:
loc and iloc:
1) index-based selection (iloc)
2) lable based selection (loc)

- Specially used for more advanced operations.

## Important:
iloc uses the Python stdlib indexing scheme, where the first element of the range is included and the last one excluded. So 0:10 will select entries 0,...,9. loc, meanwhile, indexes inclusively. So 0:10 will select entries 0,...,10

```python
ipl_df.iloc[0]
```

Note: Both loc and iloc are row-first, column-second. This is the opposite of what we do in native Python, which is column-first, row-second.

This means that it's marginally easier to retrieve rows, and marginally harder to get retrieve columns.

```python
# to access columns using iloc:
ipl_df.iloc[:,4]
```

```python
# indexing with lists:
ipl_df.iloc[[1,2,5],5:]
```

```python
# indexing by negative numbers:
ipl_df.iloc[-5:,:] # last five records
```
